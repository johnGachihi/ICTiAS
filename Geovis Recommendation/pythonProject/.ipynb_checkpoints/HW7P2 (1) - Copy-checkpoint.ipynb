# Dataloader
For this part, we will start by building a custom dataloaders that inherts from the `ImageFolder` object from `torchvision`. Since the training process for the two different loss functions treated in the problem take in different input modes (a pair of images for contrastive loss, a triplet of images for triplet loss), you are asked to use two different dataloaders.

`PairLoader`is completed for you already; it forms pairs of two images in the dataset, which can be from the same dog or different dogs. During training, the `__getitem__` function is called and it returns an anchor image, a comparison image, and a value indicating whether they are from the same dog (1 for yes, 0 for no).

You are tasked to complete `TripletLoader`, which should load triplets, that is, three images where the first image is the anchor image, the second is a positive case (i.e., from the same dog), and the third is a negative case (i.e. from a different dog).

```python
# NO NEED TO MODIFY THIS BLOCK
class PairLoader(torchvision.datasets.ImageFolder):
    def __init__(
            self,
            root: str,
            transform = None,
            target_transform = None,
            is_valid_file = None,
    ):
        super(PairLoader, self).__init__(root=root,
                                               transform=transform,
                                               target_transform=target_transform,
                                               is_valid_file=is_valid_file)
        self.pairs = self.get_pairs()

    def get_pairs(self):
        pairs = []
        labels = list(self.class_to_idx.values())
        targets = np.array(self.targets)
        pos = [np.argwhere(targets == labels[i])[[0, -1]] for i in range(len(labels))]
        neg = [np.argwhere(targets != labels[i])[[0, -1]] for i in range(len(labels))]
        for i in range(len(labels)):
            i_a, i_p = np.random.choice(np.arange(pos[i][0], pos[i][1]+1), 2, replace=False)
            i_n = np.random.choice(np.arange(neg[i][0], neg[i][1]+1), 1)[0]
            pairs.extend([[i_a, i_p, 1], [i_a, i_n, 0]])

        return np.array(pairs)

    def __len__(self):
            return self.pairs.shape[0]

    def __getitem__(self, index):
        i = self.pairs[index]
        a, a_target = self.samples[i[0]]
        c, c_target = self.samples[i[1]]
        img_a = self.loader(a)
        img_c = self.loader(c)

        if self.transform is not None:
            img_a = self.transform(img_a)
            img_c = self.transform(img_c)

        return img_a, img_c, i[2]
```

```python
# COMPLETE THIS BLOCK; YOU MAY MODIFY THIS CLASS HOWEVER YOU SEE FIT, AS LONG AS YOU KEEP THE NAME TripletLoader
class TripletLoader(torchvision.datasets.ImageFolder):
    def __init__(
            self,
            root: str,
            transform = None,
            target_transform = None,
            is_valid_file = None,
    ):
        super(TripletLoader, self).__init__(root=root,
                                               transform=transform,
                                               target_transform=target_transform,
                                               is_valid_file=is_valid_file)
        self.triples = self.get_triple()

    def get_triple(self):
        triples = []
        labels = list(self.class_to_idx.values())
        targets = np.array(self.targets)
        pos = [np.argwhere(targets == labels[i]).flatten() for i in range(len(labels))]
        neg = [np.argwhere(targets != labels[i]).flatten() for i in range(len(labels))]
        for i in range(len(labels)):
            i_a, i_p = np.random.choice(np.array(pos[i]), 2, replace=False)
            i_n = np.random.choice(neg[i], 1)[0]
            triples.extend([[i_a, i_p, i_n]])
        # print(triples)
        return np.array(triples)

    def __len__(self):
        return self.triples.shape[0]

    def __getitem__(self, index):
        i = self.triples[index]
        a, a_target = self.samples[i[0]]
        p, p_target = self.samples[i[1]]
        n, n_target = self.samples[i[2]]
        img_a = self.loader(a)
        img_p = self.loader(p)
        img_n = self.loader(n)

        if self.transform is not None:
            img_a = self.transform(img_a)
            img_p = self.transform(img_p)
            img_n = self.transform(img_n)

        return img_a, img_p, img_n
```

# Model
In this part, you are tasked with designing a model. Given the task involves images, start by implementing a two-layer CNN plus a final linear layer and try improving the model from there. The images are 160x160 pixels, with 3 colour channels.

For upgrade ideas, try making a deeper CNN or implementing elements of ResNet. You only need to submit your best model design. Keep in mind that more complex models will take longer to train.

For submission to the autograder, copy the code in the block below to the file `model.py`.

```python
class DogModel(nn.Module):
    def __init__(self):
        super(DogModel, self).__init__()

        mobilenet = torchvision.models.mobilenet_v3_small(weights='DEFAULT')
        mobilenet.requires_grad_(False)

        self.model = nn.Sequential(
            mobilenet,

            nn.Linear(1000, 2048),
            nn.Hardswish(),
            nn.Dropout(0.2),

            nn.Linear(2048, 1000)
        )


    def forward(self, x, evalMode=False):
        return self.model(x)
```

# Custom Loss Function
`ContrastiveLoss` is completed for you already. You are tasked to complete `TripletLoss`, which should work with the `TripletLoader`class you made earlier.

```python
# NO NEED TO MODIFY THIS BLOCK
class ContrastiveLoss(nn.Module):
    def __init__(self):
        super(ContrastiveLoss, self).__init__()
        self.margin = 0.2
        self.distance = nn.CosineSimilarity()

    def forward(self, anchor, compare, identical):
        d = self.distance(anchor, compare)
        loss = d * identical + (1 - identical) * (d - self.margin - d)
        return loss.mean()
```

```python
# COMPLETE THIS BLOCK; YOU MAY MODIFY THIS CLASS HOWEVER YOU SEE FIT, AS LONG AS YOU KEEP THE NAME TripletLoss
class TripletLoss(nn.Module):
    def __init__(self, alpha=0.2):
        super(TripletLoss, self).__init__()

        self.loss = nn.TripletMarginLoss()

    def forward(self, anc, pos, neg):
      return self.loss(anc, pos, neg)
```

# Training Loop
You will need to write your own training loop and validation process

```python
#WRITE YOUR VALIDATION FUNCTION HERE; YOU MAY CHANGE THE FUNCTION AS YOU SEE FIT AS LONG AS YOU KEEP THE SAME NAME
def validate(model, test_loader, test_criterion):
    model.eval() # TURNING OFF TRAINING MODE SO WEIGHTS DON'T CHANGE

    truths = []
    scores = []

    with torch.no_grad():
      for i, (img_a, img_c, labels) in enumerate(test_loader):
          img_a, img_c, labels = img_a.to(device), img_c.to(device), labels.to(device)
          emb_a = model(img_a)
          emb_c = model(img_c)
          similarity_score = nn.CosineSimilarity()(emb_a, emb_c).cpu()

          truths.extend(labels.tolist())
          scores.extend(similarity_score.tolist())

    model.train() # TURNING TRAINING MODE BACK ON

    auc_score = metrics.roc_auc_score(truths, scores)
    return auc_score
```

```python
# INITIALIZING DATA LOADERS, YOU MAY MODIFY THIS BLOCK AS YOU SEE FIT
train_dataset_p = PairLoader(root='train',
                                   transform=torchvision.transforms.ToTensor())
train_dataloader_p = torch.utils.data.DataLoader(train_dataset_p, batch_size=64,
                                               shuffle=True, num_workers=0)

train_dataset_t = TripletLoader(root='train',
                                   transform=torchvision.transforms.ToTensor())
train_dataloader_t = torch.utils.data.DataLoader(train_dataset_t, batch_size=64,
                                               shuffle=True, num_workers=0)

dev_dataset = PairLoader(root='validation',
                                               transform=torchvision.transforms.ToTensor())
dev_dataloader = torch.utils.data.DataLoader(dev_dataset, batch_size=64,
                                             shuffle=True, num_workers=0)
```

## Training with contrastive loss

```python
#WRITE YOUR TRAINING LOOP FUNCTION HERE; YOU MAY CHANGE THE FUNCTION AS YOU SEE FIT AS LONG AS YOU KEEP THE SAME NAME
def train(model, data_loader, test_loader, train_criterion, test_criterion, optimizer, scheduler=None, n_epochs=1, done_epochs=0):
    model.train() # SETTING MODEL TO TRAIN MODE

    for epoch in range(n_epochs):

        for batch_num, (img_a, img_c, label) in enumerate(data_loader):
            # INNER LOOP FOR EPOCH
            img_a, img_c, label = img_a.to(device), img_c.to(device), label.to(device)

            optimizer.zero_grad()

            img1_out = model(img_a)
            img2_out = model(img_c)

            loss = train_criterion(img1_out, img2_out, label)
            loss.backward()

            optimizer.step()

        scheduler.step()

        val_score = validate(model, test_loader, test_criterion)
        print('{} Validation Score: {:.4f}. lr {:.4f}'.format(epoch, val_score, optimizer.param_groups[0]["lr"]))
```

```python
# INITIALIZING THE MODEL AND CALLING THE TRAINING LOOP; YOU MAY CHANGE THIS BLOCK AS YOU SEE FIT
learningRate = 0.01
weightDecay = 0.0

model = DogModel()

test_criterion = nn.CosineSimilarity()
train_criterion = ContrastiveLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learningRate, weight_decay=weightDecay, momentum=0.9)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

# model.train()
model.to(device)
train(model, train_dataloader_p, dev_dataloader, train_criterion, test_criterion, optimizer, scheduler, n_epochs=40)
```

## Training with Triplet Loss

```python
def t_train(model, data_loader, test_loader, train_criterion, test_criterion, optimizer, scheduler=None, n_epochs=1, done_epochs=0):
    model.train()

    for epoch in range(n_epochs):

        for batch_num, (img_a, img_p, img_n) in enumerate(data_loader):
            # INNER LOOP FOR EPOCH
            img_a, img_p, img_n = img_a.to(device), img_p.to(device), img_n.to(device)

            optimizer.zero_grad()

            img_a_out = model(img_a)
            img_p_out = model(img_p)
            img_n_out = model(img_n)

            # print("Train_t", (img_a_out==img_p_out).all())

            loss = train_criterion(img_a_out, img_p_out, img_n_out)
            loss.backward()

            optimizer.step()

        scheduler.step()

        val_score = validate(model, test_loader, test_criterion)
        print('{} Validation Score: {:.4f}. lr {}'.format(epoch, val_score, optimizer.param_groups[0]["lr"]))

t_model = DogModel()

test_criterion = nn.CosineSimilarity()
train_criterion = TripletLoss()
optimizer = torch.optim.SGD(t_model.parameters(), lr=0.01, weight_decay=weightDecay, momentum=0.5)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

t_model.to(device)
t_train(t_model, train_dataloader_t, dev_dataloader, train_criterion, test_criterion, optimizer, scheduler, n_epochs=40)
```

# Testing Model

Here, you are provided a dataloader for the test set. You will need to implement a subroutine that loops through the test set and compares pairs retrieved from the loader in order to compute scores for obtaining the AUC ROC. Please use `roc_auc_score` in the `sklearn` package to compute the score and use the `sklearn` and `matplotlib` libraries to generate graphs.

The test set is similar to the test set we use to automatically grade. Since the test sets are small, there is a chance they differ by a few percentage points; if there is a significant discrepancy, you may try running training again to have a more consistent result.

```python
# DO NOT MODIFY THIS PART
class TestLoader(torchvision.datasets.ImageFolder):
    def __init__(
            self,
            root: str,
            transform = None,
            target_transform = None,
            is_valid_file = None,
    ):
        super(TestLoader, self).__init__(root=root,
                                               transform=transform,
                                               target_transform=target_transform,
                                               is_valid_file=is_valid_file)
        self.pairs = self.get_pairs()

    def get_pairs(self):
        pairs = []
        labels = list(self.class_to_idx.values())
        targets = np.array(self.targets)
        pos = [np.argwhere(targets == labels[i])[[0, -1]] for i in range(len(labels))]
        for i in range(len(labels)):
            i_a, i_p = np.random.choice(np.arange(pos[i][0], pos[i][1]+1), 2, replace=False)
            pairs.append([i_a, i_p, int(i >= 100)])
        return np.array(pairs)

    def __len__(self):
            return self.pairs.shape[0]

    def __getitem__(self, index):
        i = self.pairs[index]
        a, a_target = self.samples[i[0]]
        c, c_target = self.samples[i[1]]
        img_a = self.loader(a)
        img_c = self.loader(c)

        if self.transform is not None:
            img_a = self.transform(img_a)
            img_c = self.transform(img_c)

        return img_a, img_c, i[2]
```

```python
# LOADING DATASETS
test_dataset = TestLoader(root='test',
                        transform=torchvision.transforms.ToTensor())
test_dataloader = torch.utils.data.DataLoader(test_dataset, batch_size=64,
                                             shuffle=False, num_workers=0)
```

```python
def verify(test_loader, model):
    model.eval()
    metric = nn.CosineSimilarity()
    scores = []
    truths = []

    for batch_num, (image_a, image_c, labels) in enumerate(test_loader):
        image_a = image_a.to(device)
        image_c = image_c.to(device)

        output_a = model(image_a)
        output_c = model(image_c)

        score = metric(output_a, output_c).cpu()
        scores.extend(score.tolist())
        truths.extend(labels.tolist())
    return np.asarray(truths), np.asarray(scores)
```

## Performance of Model trained with Contrastive Loss

```python
truths, scores = verify(test_dataloader, model)
auc = metrics.roc_auc_score(truths, scores, multi_class="ovr")
print(auc)
```

```python
fig, ax = plt.subplots()
fpr, tpr, thresholds = metrics.roc_curve(truths, scores)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='Contrastive Loss model')
display.plot(ax=ax)
ax.set_title("ROC of model trained with Contrastive Loss")
plt.show()
```

```python
class ImageLoader(torchvision.datasets.ImageFolder):
  def __init__(self):
    super(ImageLoader, self).__init__(root="validation")

  def __len__(self):
    return len(self.samples)

  def __getitem__(self, index):
    img, _ = self.samples[index]
    img = self.loader(img)
    img = torchvision.transforms.ToTensor()(img)
    return img
```
